# by laptop has 4 cores, and 8 logical processors.

```python
import matplotlib.pyplot as plt

# Sample data
x = [2, 3, 4]
y1 = [3.63994, 3.39294, 3.47362]
y2 = [28.933, 39.6556, 50.3298]
y3 = [73.5717, 120.484, 289.986]
y4 = [276.095, 260.841, 206.354]
y5 = [633.274, 765.954, 665.926]

# Plot the data
plt.plot(x, y1, marker='o', label='N = 1,000,000')
plt.plot(x, y2, marker='o', label='N = 5,000,000')
plt.plot(x, y3, marker='o', label='N = 10,000,000')
plt.plot(x, y4, marker='o', label='N = 25,000,000')
plt.plot(x, y5, marker='o', label='N = 50,000,000')

# Add labels and title
plt.xlabel('Number of threads')
plt.ylabel('Time in second')
plt.title('Execution Time')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
